## Log 数据处理

```python
import numpy as np
```

```python
file = open('data/log')
res = file.readlines()
datas = []
for re in res:
    new_line = re.split('\n')[0][1:]
    month, date, time, log_type = new_line.split(' ')[0:4]
    data = new_line.split(': ',2)[2]
    datas.append([month, date, time[0:-1], log_type[1:-1], data])
    
npdatas = np.array(datas)
npdatas.shape
ori_data = npdatas[:,4]
ori_data
```

```python
def change_data(ori_data):
    long_sent = ' '.join(ori_data).lower()
    words_list = list(set(long_sent.split()))
    print(len(words_list))
    new_datas = []
    words_count = create_counts(words_list)   
    for sent in ori_data:
        _line = []
        for word in sent.split():
            this_index = words_list.index(word.lower())
            words_count[word.lower()] += 1
            _line.append(this_index)
        new_datas.append(_line)
    ## 作业：数据长度补充，请补零 ##
    
    ################################
    
    return new_datas, words_count

def create_counts(words_list):
    words_count = {}
    for word in words_list:
        words_count[word]= 0
    return words_count
```

```python
new_datas,words_count = change_data(ori_data)
```
