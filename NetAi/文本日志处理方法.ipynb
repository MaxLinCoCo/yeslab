```python
import numpy as np

file = open('data/log')
res = file.readlines()
datas = []
for re in res:
    new_line = re.split('\n')[0][1:]
    month, date, time, log_type = new_line.split(' ')[0:4]
    data = new_line.split(': ',2)[2]
    datas.append([month, date, time[0:-1], log_type[1:-1], data])
```

```python
npdatas = np.array(datas)
npdatas.shape
```

```python
long_sent = ' '.join(npdatas[:,4])
long_sent = long_sent.lower()
word_list = list(set([word for word in long_sent.split()]))
word_list
```

```python
word_count_list = [0]*len(word_list)
num_datas = []

for raw in range(npdatas.shape[0]):
    _line = []
    for word in npdatas[raw,4].split():
        this_index = word_list.index(word.lower())
        word_count_list[this_index] += 1
        _line.append(this_index)
    num_datas.append(_line)
    
num_datas
```
