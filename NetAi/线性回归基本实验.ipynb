## 实验1

```python
A = [0,1,2,3,4]
B = [0,1,2,3]
NUM_DATA = []

for num in range(200):
    x1 = np.random.choice(A)
    x2 = np.random.choice(B)
    NUM_DATA.append([x1,x2])
    
NUM_DATA = np.array(NUM_DATA)
NUM_LABEL = 3 * NUM_DATA[:,0]**2 + 5 * NUM_DATA[:,1]**3

train_x = NUM_DATA[0:180]
train_y = NUM_LABEL[0:180]
test_x = NUM_DATA[180:]
test_y = NUM_LABEL[180:]
```

```python
model = linear_model.LinearRegression()
model.fit(train_x, train_y)
```

```python
model.score(train_x, train_y)
```

```python
model.score(test_x, test_y)
```

```python
model.predict(test_x)
```

```python
test_y
```

```python
plt_x = list(range(20))
plt.subplot(121)
plt.plot(plt_x, test_y, 'g')
plt.subplot(122)
plt.plot(plt_x, model.predict(test_x), 'r')
plt.show()
```

```python
plt_x = list(range(20))
plt.plot(plt_x, test_y, 'g')
plt.plot(plt_x, model.predict(test_x), 'r')
plt.show()
```

## 实验2

* CRIM：城镇人均犯罪率；
* ZN：住宅用地超过 25000 sq.ft. 的比例；
* INDUS：城镇非零售商用土地的比例；
* CHAS：查理斯河空变量（如果边界是河流，则为1；否则为0）；
* NOX：一氧化氮浓度；
* RM：住宅平均房间数；
* AGE：1940 年之前建成的自用房屋比例；
* DIS：到波士顿五个中心区域的加权距离；
* RAD：辐射性公路的接近指数；
* TAX：每 10000 美元的全值财产税率；
* PTRATIO：城镇师生比例；
* B：1000（Bk-0.63）^ 2，其中 Bk 指代城镇中黑人的比例。
* LSTAT：人口中地位低下者的比例。
* MEDV：自住房的平均房价，以千美元计。

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
from sklearn.datasets import load_boston

np.set_printoptions(suppress=True)
```

```python
boston = load_boston()
```

```python
boston.data.shape
```

```python
d_min = boston.data.min(0)
d_max = boston.data.max(0)

new_data = (boston.data - d_min) / (d_max - d_min)
```

```python
ratio = int(new_data.shape[0]*0.8)
train_x = new_data[0:ratio]
train_y = boston.target[0:ratio]
test_x = new_data[ratio:]
test_y = boston.target[ratio:]
```

```python
boston_model = linear_model.LinearRegression()
boston_model.fit(train_x, train_y)
print(boston_model.score(train_x, train_y))
print(boston_model.score(test_x, test_y))
```

```python
plt_x = list(range(102))
plt.plot(plt_x, test_y, 'g')
plt.plot(plt_x, boston_model.predict(test_x), 'r')
plt.show()
```
