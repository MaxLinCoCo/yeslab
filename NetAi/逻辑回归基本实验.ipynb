# 逻辑回归基本实验

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
Datas = np.array([[1.,2.],[2.,4.],[0.,5.],[3.,1.],[4.,2.],[5.,1.]])
Label = np.array([0,0,0,1,1,1])

point_x = Datas[:,0]
point_y = Datas[:,1]

def plt_points(point_x,point_y,Label):
    for index in range(Label.shape[0]):
        if Label[index] == 0:
            color = 'ro'
        else:
            color = 'gx'
        plt.plot(point_x[index], point_y[index], color)
        
plt_points(point_x,point_y,Label)
plt.show()
```

```python
model = LogisticRegression(verbose=1)
model.fit(Datas, Label)
model.score(Datas, Label)
```

```python
test_x = np.array([[1.5, 2.5],[2,2],[3.5, 3.]])
test_y = np.array([0,1,1])
result = model.predict(test_x)
```

```python
def plot_decision_boundary(model, axis):  
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]
    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9']) 
    plt.contourf(x0, x1, zz, linewidth=5, cmap=custom_cmap)
```

```python
plot_decision_boundary(model, axis=[-0.5, 6, -0.5, 6])
plt_points(point_x,point_y,Label)
plt.show()
```

```python
plt.subplot(211)
plot_decision_boundary(model, axis=[-0.5, 6, -0.5, 6])
plt_points(test_x[:,0],test_x[:,1],result)
plt.subplot(212)
plot_decision_boundary(model, axis=[-0.5, 6, -0.5, 6])
plt_points(test_x[:,0],test_x[:,1],test_y)

plt.show()
```

```python
model.score(test_x, test_y)
```

# 公开数据集上的实验

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
```

```python
iris_datas = load_iris()
train_x = iris_datas['data']
train_y = iris_datas['target']
```

```python
train_x.shape
```

```python
train_y
```

```python
model = LogisticRegression(max_iter=30)
model.fit(train_x, train_y)
print(model.score(train_x,train_y))
result = model.predict(train_x)
```

```python
def plt_points(point_x,point_y,Label):
    plt.figure(figsize=(8,8))
    for index in range(Label.shape[0]):
        if Label[index] == 0:
            color = 'ro'
        elif Label[index]== 1:
            color = 'gx'
        else:
            color = 'b+'
        plt.plot(point_x[index], point_y[index], color)

plt_x1 = train_x[:,0]
plt_x2 = train_x[:,1]
plt_x3 = train_x[:,2]
plt_x4 = train_x[:,3]
plt_x5 = train_x.mean(1)
plt_x6 = train_x.std(1)

plt_points(plt_x1,plt_x2,result)
plt_points(plt_x1,plt_x2,train_y)

plt.show()
```

```python
plt_points(plt_x3,plt_x4,result)
plt_points(plt_x3,plt_x4,train_y)

plt.show()
```

```python
plt_points(plt_x5,plt_x6,result)
plt_points(plt_x5,plt_x6,train_y)

plt.show()
```
